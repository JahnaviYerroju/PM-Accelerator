```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
q1 = data['temperature_celsius'].quantile(0.25)
q3 = data['temperature_celsius'].quantile(0.75)
IQR = q3 - q1
lower_bound = q1 - 1.5 * IQR
upper_bound = q3 + 1.5 * IQR
data = data[(data['temperature_celsius'] >= lower_bound) & (data['temperature_celsius'] <= upper_bound)]
data['temperature_normalized'] = (data['temperature_celsius'] - data['temperature_celsius'].min()) / (data['temperature_celsius'].max() - data['temperature_celsius'].min())
print(data.describe())
```

```python
df_numeric = data.select_dtypes(include=[np.number])
correlation_matrix = df_numeric.corr()

# Plotting correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

## Temperature distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['temperature_celsius'], kde=True)
plt.title('Temperature Distribution')
plt.xlabel('Temperature (°C)')
plt.show()

## Precipitation vs Temperature scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='temperature_celsius', y='precip_mm', data=data)
plt.title('Precipitation vs Temperature')
plt.xlabel('Temperature (°C)')
plt.ylabel('Precipitation (mm)')
plt.show()

## Box plot of temperature by continent
continent_mapping = {
    'United States': 'North America', 'Canada': 'North America', 
    'Brazil': 'South America', 'Argentina': 'South America', 
    'United Kingdom': 'Europe', 'France': 'Europe', 'Germany': 'Europe', 
    'China': 'Asia', 'India': 'Asia', 'Japan': 'Asia', 
    'Australia': 'Australia', 'New Zealand': 'Australia', 
    'Egypt': 'Africa', 'South Africa': 'Africa', 'Nigeria': 'Africa'
}

data['continent'] = data['country'].map(continent_mapping)

# Handle missing continent values
data['continent'] = data['continent'].fillna('Other')

plt.figure(figsize=(12, 6))
sns.boxplot(x='continent', y='temperature_celsius', data=data)
plt.title('Temperature Distribution by Continent')
plt.xlabel('Continent')
plt.ylabel('Temperature (°C)')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

data['last_updated'] = pd.to_datetime(data['last_updated'])

data = data.sort_values('last_updated')
data.set_index('last_updated', inplace=True)

target_variable = 'temperature_celsius'

train_size = int(len(data) * 0.8)
train, test = data[:train_size], data[train_size:]

model = ARIMA(train[target_variable], order=(1,1,1))
results = model.fit()

predictions = results.forecast(steps=len(test))

mse = mean_squared_error(test[target_variable], predictions)
mae = mean_absolute_error(test[target_variable], predictions)
r2 = r2_score(test[target_variable], predictions)

print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
print(f"R-squared Score: {r2}")

plt.figure(figsize=(12, 6))
plt.plot(train.index, train[target_variable], label='Training Data')
plt.plot(test.index, test[target_variable], label='Testing Data')
plt.plot(test.index, predictions, label='Predictions')
plt.title(f'ARIMA Forecast for {target_variable}')
plt.xlabel('Date')
plt.ylabel(target_variable)
plt.legend()
plt.show()
```
